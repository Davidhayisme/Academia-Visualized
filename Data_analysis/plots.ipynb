```python
#Diseaase bar graph, x-axis is disease, y-axis is the authors
#author plot , static 
import matplotlib.pyplot as plt
import numpy as np

# Sample data:  A dictionary is a good way to organize this
disease_authors = {
    "COVID-19": 125,
    "Influenza": 80,
    "Malaria": 95,
    "Tuberculosis": 110,
    "HIV/AIDS": 130,
    "Diabetes": 70,
    "Cancer": 150,
    "Heart Disease": 140,
    "Asthma": 60,
    "Alzheimer's": 50,
}

# Extract data for plotting.  Important to maintain order.
diseases = list(disease_authors.keys())
author_counts = list(disease_authors.values())

# Create the bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(diseases, author_counts, color='skyblue')  # Use a named color

# Add labels and title
plt.xlabel("Disease", fontsize=12)
plt.ylabel("Number of Authors", fontsize=12)
plt.title("Number of Authors per Disease", fontsize=14)

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and align x-ticks

# Add gridlines for easier comparison
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on the bars.  This makes the graph more informative.
for i, count in enumerate(author_counts):
    plt.text(diseases[i], count + 3, str(count), ha='center', va='bottom', fontsize=10) # Adjust vertical offset (count + 3) as needed

# Adjust layout to prevent labels from being cut off
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data:  Dictionary where keys are diseases, and values are *lists* of publications per year.
#this is static
disease_yearly_publications = {
    "COVID-19": [5, 10, 25, 50, 60],  # Example: Years [2020, 2021, 2022, 2023, 2024]
    "Influenza": [15, 20, 18, 22, 25],
    "Malaria": [10, 12, 15, 14, 16],
    "Tuberculosis": [18, 20, 22, 21, 23],
    "HIV/AIDS": [20, 25, 30, 28, 32],
}

# Years (assuming a fixed range for this example)
years = [2020, 2021, 2022, 2023, 2024] #important to define years

# Create the point plot
plt.figure(figsize=(10, 6))

# Iterate through each disease and plot its data
for disease, publications in disease_yearly_publications.items():
    plt.plot(years, publications, marker='o', label=disease, linestyle='-')  # 'o' for points, '-' for line

# Add labels and title
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Publications", fontsize=12)
plt.title("Yearly Publications per Disease", fontsize=14)

# Add legend to distinguish diseases
plt.legend(fontsize=10)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Ensure only whole numbers are shown on the x-axis
plt.xticks(years)  # <--- Added this line

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
